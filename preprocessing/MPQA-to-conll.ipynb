### Scheme
B - beginning
I - in
O - nothing
S - singleton
E - end
each token is labeled with two layers: modal and prejacent, like so:

```[{'sentence_number': 1, 'token_number': 1, 'token': 'Private', 'modal-BIOSE': 'O', 'prejacent': 'B'}]```


### problems to tackle
1. the sentence has more than one prejacent/modal.
2. The modal target token appears more than once in the sentence (whether as a modal or not)
3. The modal appears within the prejacent (about half the times)
4. The prejacent is unknown (appears in the df as '[]')
5. The modal target consists of more than one word.

### approach
1. First get sentences where there is only one modal and the prejacent contains the modal.

````python
def insert_element(element: str) -> list:
    taglist = []
    if len(element.split()) == 1:
        taglist.append('S')
    elif len(element.split()) > 1:
        taglist.append('B')
        [taglist.append('I') for e in element.split()[1:-1]]
        taglist.append('E')
    return taglist

def insert_sentence_with_one_modal_not_contained_in_prejacent(sentence: str, modal: str, prejacent: str) -> list:
    modals = []
    prejs = []
    sent_replaced = sentence.replace(modal, ' ~M~ ')
    sent_replaced = sent_replaced.replace(prejacent, ' ~P~ ')
    for token in sent_replaced.split():
        if token == '~M~':
            modals.append(insert_element(modal))
        else:
            modals.append('O')
        if token == '~P~':
            prejs.extend(insert_element(prejacent))
        else:
            prejs.append('O')
            
    
    return modals, prejs

def insert_sentence_with_one_modal_contained_in_prejacent(sent_no: int, sentence: str, modal: str, prejacent: str) -> list:
    """Since a split of a prejacent can appear several times, it is necessary to number them"""
    tags = []
    sent_replaced = sent.replace(prejacent, '@P@')
#     split_prej = prejacent.split(modal)
#     for x in range(len(split_prej)):
#         prejacent = prejacent.replace(split_prej[x], ' ~{}P~ '.format(str(x)))
    prejacent = prejacent.replace(modal, ' ~M~ '.format(str(x)))
    for token in sent_replaced.split():
        if token != '@P@':
            row = {'sentence_number': sent_no, 'token_number': 1, 'token': 'Private', 'modal-BIOSE': 'O', 'prejacent': 'B'}
            tags.append('{}###{}'.format(token, 'O'))
        else:
            for p in prejacent.split():
                if p == '~M~':
                    tags.extend(insert_element(modal, 'modal'))
                else:
                    tags.extend(insert_element(modal, 'modal'))



        
````

3. in xxx.mark.xml selecet all rows which are structured as `<mark id="modal_1" xlink:href="#sTok1"/>`. If the modal contains more than one token, it appears as `<mark id="modal_1" xlink:href="#sTok1 #sTok2"/>`. The prejacent appears the same but with `id="other_1"`.

````python
def tag_BIOSE(token_list, col):
    for token_group in token_list:
        if len(token_group) == 1:
            df.at[df['ix'] == int(token_group[0]), col] = 'S'
        else:
            df.at[df['ix'] == int(token_group[0]), col] = 'B'
            df.at[df['ix'] == int(token_group[-1]), col] = 'E'
            for token in token_group[1:-1]:
                df.at[df['ix'] == int(token), col] = 'I'
````
